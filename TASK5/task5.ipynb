```python
# Import all the necessary library 
import pandas as pd
```

```python
#add the data 
main = pd.read_csv("train.csv")
submission = pd.read_csv("gender_submission.csv")
```

```python
main.head(10)
```

```python
submission.head(10)
```

```python
#merge the data left join
final = pd.merge(main,submission,how='left',on='PassengerId')
final
```

```python
#delete the null column 
del final['Survived_y']
```

```python
final.head(10)
```

```python
final.rename(columns={"PassengerId":"passengerid","Survived_x":"survived","Pclass":"pclass","Name":"name","Sex":"sex","Age":"age","SibSp":"sibsp","Parch":"parch","Ticket":"ticket","Fare":"fare","Cabin":"cabin","Embarked":"embarked"},inplace=True)
```

```python
final.head(10)
```

```python
final.describe()
```

```python
survivors_by_gender = final[final['survived'] == 1].groupby('sex').size()
print(survivors_by_gender)
```

```python
final_age = final.dropna(subset=['age'])

# Define age bins and labels
bins = [0, 20, 40, 60, 80, 100]
labels = ['(0-20)', '(21-40)', '(41-60)', '(61-80)', '(81-100)']

# Create a new column for age group
final_age['age_group'] = pd.cut(final_age['age'], bins=bins, labels=labels, include_lowest=True)

# Count the number of survivors in each age group
survivors_by_age_group = final_age[final_age['survived'] == 1].groupby('age_group').size()

print(survivors_by_age_group)
```

```python
survivors_by_pclass = final[final['survived'] == 1].groupby('pclass').size()
survivors_by_pclass
```

```python
final.to_csv("titanic.csv")
```

```python
sns.pairplot(final[['survived', 'pclass', 'age', 'fare']].dropna(), hue='survived')
plt.show()
```

**Observation:**

* People who paid **higher fares** (upper right in Fare axis) had **higher survival**.
* Passengers in **lower classes (higher Pclass number)** often did **not survive**.
* No strong relationship between `Age` and `Survival`, but many younger people did survive.

```python
sns.heatmap(final.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

**Observation:**

* `Fare` and `Pclass` show a **negative correlation** (higher fare → lower Pclass number).
* `Survived` correlates **positively** with `Fare` and **negatively** with `Pclass` (class 1 passengers more likely to survive).
* Weak correlation with `Age` and `SibSp`.

```python
sns.histplot(final['age'].dropna(), kde=True)
plt.title('Distribution of Age')
plt.show()
```

**Observation:**

* Age distribution is **right-skewed**.
* Most passengers were between **20 and 40 years** old.
* Very few elderly passengers.

```python
sns.boxplot(x='fare', data=final)
plt.title('Boxplot of Fare')
plt.show()
```

**Observation:**

* **Fare** has many **outliers** — a few passengers paid very high fares.
* Median fare is relatively low.
* Fare distribution is **highly skewed**.

```python
sns.countplot(x='survived', hue='sex', data=final)
plt.title('Survival Count by Gender')
plt.show()
```

**Observation:**

* **Females** were much more likely to **survive** than males.
* Among non-survivors, most were **males**.

```python
sns.countplot(x='survived', hue='pclass', data=final)
plt.title('Survival Count by Class')
plt.show()
```

**Observation:**

* Most survivors were from **1st class**.
* Passengers in **3rd class** had the **lowest survival rate**.

```python
sns.scatterplot(x='age', y='fare', hue='survived', data=final)
plt.title('Age vs Fare (colored by Survival)')
plt.show()
```

**Observation:**

* Many survivors paid **higher fares**.
* Younger people appear across survival outcomes.
* Some **elderly** survived, especially those who paid high fares.


<h1>Summary</h1>



 Gender and class had a strong impact on survival. Females and passengers in 1st class had the highest chances of survival.

 Most passengers were between 20–40 years old. Age did not strongly determine survival on its own.

 Higher-paying passengers had better survival rates, likely due to their access to first-class cabins.

 Survival decreased from 1st to 3rd class. Class was a strong predictor of outcome.

 Fare and class were the most correlated features with survival; age and siblings/spouses were less impactful.
