```python
# Import all the necessary library 
import pandas as pd
```

```python
# Reading the file 
df = pd.read_excel("CHOCOLATE DATASET.xlsx")
```

```python
# Fetching the top 15 rows to check the dataset 
df.head(15)
```

```python
# Fetching the last 5 rows to check the dataset 
df.tail()
```

```python
"""
Identifying the null value: 
we have found no null values in the columns
"""
df.isna().sum()
```

```python
"""
Identifying the duplicate values:
there is no duplicate value detected 
"""
df.drop_duplicates(inplace=True)
```

```python
# Identifying the shape of the dataframe
df.shape
```

```python
# Renaming the column into the lower case and removing the unnecessary space
df.rename(columns={"Sales Person":"sales_person","Country":"country","Product":"product","Date":"date","Amount":"amount","Boxes Shipped":"boxes_shipped"},inplace=True)
```

```python
# Detecting how the values are stored in the categorical column product
df["product"].value_counts()
```

```python
# Detecting how the values are stored in the categorical column country
df["country"].value_counts()
```

```python
# Getting an idea that what are data-type of column we have
df.dtypes
```

```python
# changing the date to the Format to 'dd-mm-yyyy'
df['date'] = df['date'].dt.strftime('%d-%m-%Y')
```

```python
# Fixing the amount column as the column contains the value $5,320 to 5320 
df["amount"] = df["amount"].str.replace('$','')
df["amount"] = df["amount"].str.replace(',','')
```

```python
df.head()
```

```python
# Converting the column data-type in the necessary format

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df["amount"] = df["amount"].astype(int)
```

```python
# Describing the complete dataframe to identify that how the numerical values are stored like min,max and std
df.describe()
```
